```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img_center = cv2.imread('ImageCollageData/center.jpeg')
img_bl = cv2.imread('ImageCollageData/bottom_left.jpg')
img_br = cv2.imread('ImageCollageData/bottom_right.jpg')
img_tl = cv2.imread('ImageCollageData/top_left.jpg')
img_tr= cv2.imread('ImageCollageData/top_right.jpg')
print(img_center.shape, img_bl.shape, img_br.shape,img_tl.shape,img_tr.shape)
```

```python
# Resize the shape to 200*200
img_bl = cv2.resize(img_bl, (200,200) )
img_center = cv2.resize(img_center, (100,100) )
img_br = cv2.resize(img_br, (200,200) )
img_tl = cv2.resize(img_tl, (200,200) )
img_tr = cv2.resize(img_tr, (200,200) )
print(img_center.shape, img_bl.shape, img_br.shape,img_tl.shape,img_tr.shape)
```

```python
# Check the image
plt.imshow(img_center)
```

```python
img_center = cv2.cvtColor(img_center,cv2.COLOR_BGR2RGB)

img_bl = cv2.cvtColor(img_bl,cv2.COLOR_BGR2RGB)

img_br = cv2.cvtColor(img_br,cv2.COLOR_BGR2RGB)

img_tl = cv2.cvtColor(img_tl,cv2.COLOR_BGR2RGB)

img_tr = cv2.cvtColor(img_tr,cv2.COLOR_BGR2RGB)

print(img_center.shape, img_bl.shape, img_br.shape,img_tl.shape,img_tr.shape)
```

```python
img_center = cv2.copyMakeBorder(img_center,10,10,10,10, cv2.BORDER_CONSTANT) # T B L R
img_bl= cv2.copyMakeBorder(img_bl,5,10,10,5,cv2.BORDER_CONSTANT) # T B L R
img_br= cv2.copyMakeBorder(img_br,5,10,5,10,cv2.BORDER_CONSTANT) # T B L R
img_tl= cv2.copyMakeBorder(img_tl,10,5,10,5,cv2.BORDER_CONSTANT) # T B L R
img_tr= cv2.copyMakeBorder(img_tr,10,5,5,10,cv2.BORDER_CONSTANT) # T B L R
print(img_center.shape, img_bl.shape, img_br.shape,img_tl.shape,img_tr.shape)
```

```python
# Use the NumPy Stack

# we will use numpy stack function
Horizontal1=np.hstack([img_tl,img_tr])
Horizontal2=np.hstack([img_bl,img_br])

# Now the horizontal attachment is done
Vertical_Attachment=np.vstack([Horizontal1,Horizontal2])

# Show the final attachment
plt.imshow(Vertical_Attachment)
print(Vertical_Attachment.shape)
```

```python
x_offset = 155
y_offset = 155
x_end = x_offset + img_center.shape[1]
y_end = y_offset + img_center.shape[0]
```

```python
Vertical_Attachment[y_offset:y_end,x_offset:x_end] = img_center
plt.imshow(Vertical_Attachment)
print(Vertical_Attachment.shape)
```

```python
import pandas as pd
imageMat = Vertical_Attachment
# if image is colored (RGB)
if(imageMat.shape[2] == 3):
    
  # reshape it from 3D matrice to 2D matrice
  imageMat_reshape = imageMat.reshape(-1, imageMat.shape[2])
print("Reshaping to 2D array:", 
        imageMat_reshape.shape)
```

```python
# converting it to dataframe.
mat_df = pd.DataFrame(imageMat_reshape)
  
# exporting dataframe to CSV file.
header_list = ['r' , 'g' ,'b']
mat_df.to_csv('ImgValRes1.csv',
              header = header_list,
              index = None)
```
